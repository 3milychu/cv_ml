## planes

```python
f = open('planes.pkl', 'rb')
plane_images = pickle.load(f)  # variables come out in the order you put them in
f.close()

print(plane_images.shape)

for plane in plane_images:
    bool_plane = img_as_bool(plane)
    plt.figure()
    io.imshow(bool_plane)
```
